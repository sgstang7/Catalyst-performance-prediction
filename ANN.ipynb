```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
```

```python
# 读入数据
data = pd.read_excel('features.xlsx')
```

```python
# 划分输入和输出
x = []
for col in data.columns:
    if col not in ['CO Faradaic efficiency/ %','Current Density']:
        x.append(col)
x = np.array(data[x])
y = data['CO Faradaic efficiency/ %']
```

```python
# 按照属性标号处理名词性属性
x = x.T
for i in range(x.shape[0]):
    x[i] = LabelEncoder().fit_transform(x[i])
x = x.T
```

```python
# 数据预处理
scaler1 = StandardScaler()
x = scaler1.fit_transform(x)
scaler2 = StandardScaler()
y = scaler2.fit_transform(np.array([y]).T)
```

```python
# 数据集划分
train_x, test_x, train_y, test_y = train_test_split(x, y,train_size=0.8)
```

```python
# Build an ANN Model
model = Sequential()
# Input layer
model.add(Dense(units=47, kernel_initializer='normal', activation='relu', input_dim=45))
# Hidden layer
model.add(Dense(units=49, kernel_initializer='normal', activation='relu'))
# Output layer
model.add(Dense(units=1, kernel_initializer='normal'))
# Setup optimizer
model.compile(optimizer='sgd', loss='mse', metrics=['mae'])
```

```python
# Train ANN model
history = model.fit(train_x, train_y, epochs=500, batch_size=20, validation_data=(test_x,test_y),verbose=2)
```

```python
# 测试ANN最终结果
y_pred = model.predict(test_x)
print(model.evaluate(test_x, test_y))
model.save('predict_model.h5')
score = model.evaluate(test_x, test_y, verbose = 2)
```

```python
model.summary()
```

```python
# 预测y 逆标准化
y_pred =  y_pred.reshape(len(y_pred),1)
y_pred = scaler2.inverse_transform(y_pred)
# 原始y逆标准化
test_y = test_y.reshape(len(test_y),1)
test_y = scaler2.inverse_transform(test_y)
# 画原始数据和预测数据图
plt.figure(figsize=(14,10))
plt.plot(list(test_y), label='real data')
plt.plot(y_pred, label='forecasting data')
plt.xlabel('epochs')
plt.ylabel('CO_Faradaic efficienty %')
plt.title('preddict VS True')
plt.legend()
plt.show()
```

```python
# 画损失值收敛图
plt.figure(figsize=(10,6))
plt.plot(history.history['loss'])
plt.title('mse')
plt.show()
```

```python
# 画mae收敛图
plt.figure(figsize=(10,6))
plt.plot(history.history['mae'])
plt.title('mean absolute error')
plt.show()
```
